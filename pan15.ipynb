15 章 奇跡のゲーム

　先物市場はゼロサムゲームである。一方の利益は他方の損失であり、その仕組みの中で市場は取引の機会と適切な価格という情報を提供している。市場が合理的な人たちで成り立っているとすると売値と買値の差の売買スプレッドを調整することで、仲介業者は事業を継続できる。しかし、日経225miniのような現実の先物市場では、99％の取引が最小の売買スプレッドで成立する。これでは理論にしたがった戦略ではうまく仲介業を展開できない。売買スプレッドが小さすぎて、逆選択在庫による損失を取り戻せない。プロの仲介業者であろうと、マーケットメイキングの継続は困難だ。これでは市場参加者の短期の収益は安定したものとはならない。これが取引を継続させるためには、効率的市場で期待される以上の費用が掛かる理由である。Black(1986)によると市場の機能を維持する費用はノイズトレーダーによって支払われている。賢い市場参加者は非緊急性取引を好み、マーケットメイカーはパッシブではなく、より積極的に取引をするアクティブ戦略へと移行しつつある。

マーケットは変幻自在

15.1 基本に立ち返る

　金融取引では価格は需給の均衡点で決まる。売り手と買い手が売買希望価格とその数量を提示し、売買条件が一致すると取引は成立する。
 
15.1.1 取引が成立する確率は低い

しかし条件の一致する相手と出会う確率は低く、場所と時間を定めて参加者は集まり、さらに、需給の不一致のもとでも取引が円滑に行われるように仲介のシステムを導入している。
 
15.1.2 取引を活発にするためには助っ人が必要

確率の低い出会いの機会を、仲介業者(マーケットメイカー)が在庫を抱えることで増やしている。マーケットメイカーはそのリスクに見合ったプレミアム(売買スプレッド)を受け取ることで、事業の継続を可能としている。
 
15.1.3 価格の動きは予測不可能 

マーケットメイカーであっても、価格の動きは予測不可能で、損失を被ることがある。価格の動きとは逆行する、在庫が積みあがってしまうことを逆選択リスクという。
 
15.1.4 多くの人は勝手気ままに取引している

一方で流動性を消費する側の市場参加者は、それぞれの事情により取引の判断を行う。流動性トレーダーと呼ばれ、多くの市場参加者はこのタイプに分類される。
 
15.1.5 マーケットメイカーの利益の源泉

正しい方向の在庫は利益を生むが、価格の動きと逆向きの在庫は損失を生む。正しい在庫は、反対側に置いた指値注文により閉じられ多くは売買スプレッドの1ティック(5円)を獲得して終わる。
　一方で、価格が在庫と反対方向に動き出すと、逆向きの在庫が積みあがり、損失が膨らみ続ける。同じ方向に継続して価格が動く確率は必ずある。この損失は、理論的には最良買気配で買い、それよりもわずかに高い最良売気配で売るマーケットメイクの業務による収益でまかなわれる。
　市場にマーケットメイカーと流動性トレーダーしかいないとなると、マーケットメイカーの利益は流動性トレーダーの損失であり、その逆も成り立つ。

②　取引の基本：成行と指値注文

　取引は、指値と成行という２つの基本的な注文を用いて行われる。指値注文とは売買の価格と量を提示する注文であり、指定した価格(呼値)かそれよりも良い価格で成立する。約定は約束されていないが、取り消されるまで有効である。一方、成行注文では取引数量を提示するが価格は指定しない。最良の買呼値、または売呼値などで約定する。約定価格は指定できないが、取引の成立を優先させたい場合に用いられる。

　表15.1では最良の気配値は10005円と10000円である。10005円で買うか、10000円で売ることができる。買いの成行注文を出すと、10005円か10010円で取引が成立する可能性がある。同様に売りの成行注文を出すと、10000円か9995円で取引が成立する可能性がある。可能性があるという意味は売気配・買気配で示されている数値はいつも変化していて、買いの成行注文を出した瞬間に最良売指値が消えてしまうこともしばしばだからだ。買指値注文も同様である。また、10005円に10枚の売指値を入れたとしよう。先物価格が下落してしまえば、いつまでたっても約定はしない。一方でその瞬間に売気配のすべてがキャンセルされて、10005円の売りが10枚成立してしまうこともある。売りたい場合に取引を確実に成立させたければ、10000円の買指値を狙って、10000円の売り指値を1枚入れれば大体成立する。このような注文を即時約定指値注文という。

15.2.4 損切戦略の利用(継続型)

15.2.5 マーケットメイカーの注文キャンセルモデル

　値動きの流れに頼る戦略やそれに損切りを加えた戦略では継続可能な流動性供給戦略を構築することはできない。そこで指値の売り注文を出したあとに、逆選択在庫を抱えないように、また、利食いの確率を上げるように売買のタイミングを見定め、状況が不利と判断されれば注文をキャンセルするという、指値注文が約定されるタイミングをより慎重に選ぶという戦略に切り替えてみよう。指値の発注時にはなるべく早く約定されるように待ち行列が短いものを選ぶ。その注文の約定が近くなって来た際には、その後にどれくらい長い売気配の行列があるかを確認する。また、売り注文の約定と同時に買い注文を出すのであるが、その注文がどれくらい速やかに約定されそうかを買気配の行列の長さで判断する。板の状態が不利だと思えば、売り注文をキャンセルする。

　では実際のデータを使ったバックテストの結果を見てみよう。

```python
plotanalysis(I3.pl.cumsum(),'損益','図15.9 トイモデル+注文キャンセル',
             I3.no,'約定数','図15.10　約定数 ')
print(f"総損益{I3.pl.sum():2.0f},総約定数{I3.no.sum():2.0f},\
キャンセルの総数{I3.no0.sum():2.0f} 逆選択在庫の解消{I3.sl.sum()}")
```

　図15.7は累積の損益、図15.8は取引の数である。モデルでは最良買気配と最良売気配の枚数、そして指値の注文を出した時点でその前の注文が何枚あり、それがどのように減って行き、そして時間の経過と共に最良買気配と売気配の枚数が何枚になっているのかを観察し、自分の位置の有利、不利を見極め、不利だと思えば注文を取り消し、新たな売り注文を2ティック上に置く。このシミュレーションは後知恵で最適化してあるので注意してほしい。また、板情報を使って取引戦略を組んであるが、最良買気配と最良売気配しか使っていない。これは最良気配外の指値情報を使ったとしても、大きな影響はないだろうと考えるからである。板情報自体は、約定させる気のない指値注文を提示するスプーフィング等の相場操縦の手口により、必ずしも信頼できるとは考えていないからである。また、実際には取引のスピードに応じてモデルの変数を変更するべきところを単純化してある。そのために取引のスピードが急激に上がるチャイニーズブラックマンデー以降に取引高が増えている。

　実際に収益に影響を与える要素は何であろうか？

1.	売(買)指値の水準
2.	売(買)指値の取り消しのタイミング

の2つである。売指値を頻繁に発注すると売気配値の待ち行列の後ろに常に位置し、行列の前に行くことなく注文の約定は価格水準の上昇によって引き起こされてしまう。今ある水準よりも離れたところに売指値を置くと、約定までには時間が掛かるが、時間の経過と注文の取り消しが徐々に起こり自分の注文の位置が徐々に前に移動していく。より良い待ち行列の位置に来ることができる。しかし、どのような努力をしても自分の指値の位置がどこであるかは知ることができない。
　自分よりも前の売指値の注文が約定し、自分の順番が近づいてきても、同じ水準の売指値の注文が多く残っていないと、価格は1ティック上に上昇してしまうかもしれない。そうするとポジションを解消する買指値が約定する確率は下がってしまう。したがって、自分の売指値が約定しそうなときであっても自分の位置が不利であったなら、約定の前に注文を取り消す必要がある。これが究極のストップロス注文である。取引は状況が整うまで実行してはならないのだ。

　自分が有利な位置にいなければ取引は諦めろ。

```python
def zaraba_from_market(df):
    lenprice=len(df)
    p0=df.Price.iloc[0] 
    p9=df.Price.iloc[-1] 
    for k,price0 in enumerate(df.Price):#itayose open
        if price0!=p0:
            break
    price_r=df.iloc[::-1]
    for kk,price0 in enumerate(price_r.Price):#itayose close
        if price0!=p9:
            break
    return df[k:-kk].copy()#zaraba market  
```

で始まるのループは午前と午後の立会を識別し、tradingに市場データを与える。

　trading関数では立会内の取引を実行している。売指値の価格を最良売気配値として発注することから始まる。qoは待ち行列の長さで、そのときの売気配の先物の枚数である。約定のたびにその枚数は取引高の枚数(v)だけ減っていく。
 　約定は待ち行列がゼロになったときに執行されるか、価格の水準が1ティック以上上昇したときに行われる。
  　売指値が約定するとそれよりも1ティック以上下に買指値を入れる。買指値が最良気配値となった時点から待ち行列の算出を始める。買指値が約定するのは売指値と同様に待ち行列がなくなったときか、約定価格が1ティック以上下げたときである。利食いが完了すると売指値が最良売気配値に置かれる。

```python
def toy_model(tsd):
    def trading(tmp):
        zaraba=zaraba_from_market(tmp)#ザラバ
        limit_sell=zaraba.BestAsk.iloc[0]
        ib=zaraba.BBV.iloc[0]
        io=qo=zaraba.BAV.iloc[0]
        qb=-1
        shelved=1        
        pos_sell=0
        limit_buy=0
        pl=0
        pl0=0
        sss=0
        zaraba_list=zaraba.iloc[1:].to_numpy().tolist()
        for aa in zaraba_list:        
            trade=aa[0]
            bid=aa[1]
            offer=aa[2]
            if offer<bid:
                offer=bid+50
            v=aa[3] #出来高
            ib=aa[4] #最良買気配の量
            io=aa[5] #最良売気配の量
            if limit_sell==trade and pos_sell==0 and qo>0:
                #売り注文待ち行列の減少
                qo-=v
            if limit_sell==trade and pos_sell==0 and qo<=0:#注文約定
                pos_sell=limit_sell
                if bid<=pos_sell-5:
                    limit_buy=bid
                    qb=ib
                    shelved=0
                else:
                    limit_buy=pos_sell-5
                    shelved=1                
                sss+=1
            if trade>=limit_sell+5 and pos_sell==0:#価格上昇による約定
                pos_sell=limit_sell
                if bid<=pos_sell-5:
                    limit_buy=bid
                    qb=ib
                    shelved=0
                else:
                    limit_buy=pos_sell-5
                    shelved=1                
                sss+=1
            if limit_buy==trade and pos_sell!=0 and qb>0 and shelved==0:
                qb-=v#順番待ち行列の減少
            if limit_buy==trade and pos_sell!=0 and shelved==1:
                qb=v
                shelved=0#順番待ち行列への参加
            if limit_buy==trade and pos_sell!=0 and qb<=0 and shelved==0:
                #順番待ちによる約定
                pl+=pos_sell-limit_buy-2
                pos_sell=0
                limit_buy=0
                qo=io
                limit_sell=offer
            if pos_sell!=0 and trade<=limit_buy-5:
                #値動きによる約定
                pl+=pos_sell-limit_buy-2
                pos_sell=0
                limit_buy=0
                limit_sell=offer
                qo=io
        if pos_sell!=0:##立会終了時のポジション解消
            pl0=pos_sell-offer-2
            pl+=pl0
        I.append([pl,sss,sss0,pl0])
        
    da=[];I=[];sss0=[];pl0=[]
    indexdates=pd.unique(tsd.index.date)
    for indexdate,indexdate1 in zip(indexdates[:-1],indexdates[1:]):
        tmp=pd.DataFrame()
        tmp=tsd.loc[str(indexdate)+' 8:45:00+09:00':
                    str(indexdate)+' 15:15:00+09:00']
        if not tmp.empty:
            da.append(str(indexdate)+' 09:00')
            trading(tmp)
        tmp=pd.DataFrame()
        tmp=tsd.loc[str(indexdate)+' 16:30:00+09:00':
                    str(indexdate1)+' 5:30:00+09:00']
        if not tmp.empty:
            da.append(str(indexdate)+' 16:30')
            trading(tmp)
    return pd.DataFrame(I,index=da,columns=['pl','no','no0','sl'])
```

```python
def stoploss_model(tsd,sl):
    def trading(tmp):
        zaraba=zaraba_from_market(tmp)#ザラバ
        limit_sell=zaraba.BestAsk.iloc[0]
        ib=zaraba.BBV.iloc[0]
        io=qo=zaraba.BAV.iloc[0]
        qb=-1
        shelved=1        
        pos_sell=0
        limit_buy=0
        pl=0
        pl0=0
        sss=0
        sss0=0
        zaraba_list=zaraba.iloc[1:].to_numpy().tolist()
        for aa in zaraba_list:        
            trade=aa[0]
            bid=aa[1]
            offer=aa[2]
            if offer<bid:
                offer=bid+50
            v=aa[3] #出来高
            ib=aa[4] #最良買気配の量
            io=aa[5] #最良売気配の量
            if limit_sell==trade and pos_sell==0 and qo>0:
                #売り注文待ち行列の減少
                qo-=v
            if limit_sell==trade and pos_sell==0 and qo<=0:#注文約定
                pos_sell=limit_sell
                if bid<=pos_sell-5:
                    limit_buy=bid
                    qb=ib
                    shelved=0
                else:
                    limit_buy=pos_sell-5
                    shelved=1                
                stoploss=limit_sell+sl
                qb=ib #queue bid
                sss+=1
            if trade>=limit_sell+5 and pos_sell==0:#価格上昇による約定
                pos_sell=limit_sell
                if bid<=pos_sell-5:
                    limit_buy=bid
                    qb=ib
                    shelved=0
                else:
                    limit_buy=pos_sell-5
                    shelved=1                
                stoploss=limit_sell+sl
                qb=ib
                sss+=1
            if limit_buy==trade and pos_sell!=0 and qb>0 and shelved==0:
                qb-=v
            if limit_buy==trade and pos_sell!=0 and shelved==1:
                qb=v
                shelved=0#順番待ち行列への参加
            if limit_buy==trade and pos_sell!=0 and qb<=0 and shelved==0:
                pl+=pos_sell-limit_buy-2
                pos_sell=0
                limit_buy=0
                qo=io
                limit_sell=offer
            if pos_sell!=0 and trade<=limit_buy-5:
                pl+=pos_sell-limit_buy-2
                pos_sell=0
                limit_buy=0
                limit_sell=offer
                qo=io
            if pos_sell!=0 and trade>=stoploss:
                pl+=pos_sell-offer-2
                pos_sell=0
                limit_buy=0
                limit_sell=offer
                sss0+=1
        if pos_sell!=0:##立会終了時のポジション解消
            pl0=pos_sell-offer-2
            pl+=pl0
        I.append([pl,sss,sss0,pl0])
        
    da=[];I=[];sss0=[];pl0=[]
    indexdates=pd.unique(tsd.index.date)
    for indexdate,indexdate1 in zip(indexdates[:-1],indexdates[1:]):
        tmp=pd.DataFrame()
        tmp=tsd.loc[str(indexdate)+' 8:45:00+09:00':
                    str(indexdate)+' 15:15:00+09:00']
        if not tmp.empty:
            da.append(str(indexdate)+' 09:00')
            trading(tmp)
        tmp=pd.DataFrame()
        tmp=tsd.loc[str(indexdate)+' 16:30:00+09:00':
                    str(indexdate1)+' 5:30:00+09:00']
        if not tmp.empty:
            da.append(str(indexdate)+' 16:30')
            trading(tmp)
    return pd.DataFrame(I,index=da,columns=['pl','no','no0','sl'])
```

入力[10]のcancel_modelは一度発注した売り指値注文を状況が不利になった場合にキャンセルする戦略である。取引のスピードを競う理由はここにある。いくら発注のスピードが速くても、順番待ち行列の最後に着いたのでは発注スピードは意味をなさない。しかし、キャンセルにおいてはそうはならない。指値注文のキャンセルに影響を与えるのはqqo(売り注文待ち行列の長さ)とiio(最良売気配の枚数)とiib(最良買気配の枚数)、そして利食いの大きさである。2015年8月のバックテストでは利食いのティックは2であるが、2016年から2019年では7ティック、2020年から2022年では75ティックとした。

```python
def cancel_model(tsd,qqo,iio,iib):
    def trading(tmp):
        zaraba=zaraba_from_market(tmp)#ザラバ
        limit_sell=zaraba.BestAsk.iloc[0]
        ib=zaraba.BBV.iloc[0]
        io=qo=zaraba.BAV.iloc[0]
        qb=-1
        shelved=1
        pos_sell=0
        limit_buy=0
        can=0
        pl=0
        pl0=0
        sss=0
        sss0=0
        zaraba_list=zaraba.iloc[1:].to_numpy().tolist()
        for aa in zaraba_list:        
            trade=aa[0]
            bid=aa[1]
            offer=aa[2]
            if offer<bid:
                offer=bid+50
            v=aa[3] #出来高
            ib=aa[4] #最良買気配の量
            io=aa[5] #最良売気配の量
            #print("all")#,end=" ")
            if limit_sell==offer and limit_buy==0 \
                    and (qo>qqo or io<iio or ib>iib):
                can=offer+5
                limit_sell=0
                qo=0
                sss0+=1
                #print("注文キャンセル")
            if can==offer and limit_sell==0 and pos_sell==0:
                limit_sell=offer
                can=0
                qo=io
                sss0+=1
                #print("注文キャンセル約定")
            if limit_sell==trade and pos_sell==0 and qo>0:
                #売り注文待ち行列の減少
                qo-=v
            if limit_buy==trade and pos_sell!=0 and shelved==1:
                qb=v
                shelved=0#順番待ち行列への参加
            if limit_sell==trade and pos_sell==0 and qo<=0:#注文約定
                pos_sell=limit_sell
                sss+=1
                if bid<=pos_sell-5*2:
                    limit_buy=bid
                    qb=ib
                    shelved=0
                else:
                    limit_buy=pos_sell-5*2
                    shelved=1
                #print("順番による約定",sss)
            if trade>=limit_sell+5 and pos_sell==0:#価格上昇による約定
                pos_sell=limit_sell
                sss+=1
                if bid<=pos_sell-5*2:
                    limit_buy=bid
                    qb=ib
                    shelved=0
                else:
                    limit_buy=pos_sell-5*2
                    shelved=1
                #print("価格上昇による約定",sss)
            if limit_buy==trade and pos_sell!=0 and qb>0 and shelved==0:
                qb-=v
            if limit_buy==trade and pos_sell!=0 and qb<=0 and shelved==0:
                pl+=limit_sell-limit_buy-2/8#取引回数を多数と考え手数料を減らす
                pos_sell=0
                limit_buy=0
                qo=io
                limit_sell=offer
                #print('利食い順番待ち',pl)
            if pos_sell!=0 and trade<=limit_buy-5:
                pl+=limit_sell-limit_buy-2/8#取引回数を多数と考え手数料を減ら
                pos_sell=0
                limit_buy=0
                limit_sell=offer
                #print('利食い',pl)
                qo=io
        if pos_sell!=0:##立会終了時のポジション解消
            pl0=pos_sell-offer-2/8#取引回数を多数と考え手数料を減ら
            pl+=pl0
        I.append([pl,sss,sss0,pl0])
        #print(sss,pl,sl)
        
    da=[];I=[];pl0=[]
    indexdates=pd.unique(tsd.index.date)
    for indexdate,indexdate1 in zip(indexdates[:-1],indexdates[1:]):
        #print(str(indexdate))
        tmp=pd.DataFrame()
        tmp=tsd.loc[str(indexdate)+' 8:45:00+09:00':
                    str(indexdate)+' 15:15:00+09:00']
        if not tmp.empty:
            da.append(str(indexdate)+' 09:00')
            trading(tmp)
        tmp=pd.DataFrame()
        tmp=tsd.loc[str(indexdate)+' 16:30:00+09:00':
                    str(indexdate1)+' 5:30:00+09:00']
        if not tmp.empty:
            da.append(str(indexdate)+' 16:30')
            trading(tmp)
    return pd.DataFrame(I,index=da,columns=['pl','no','no0','sl'])
```
